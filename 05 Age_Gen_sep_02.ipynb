### 男性用区分

```python
split_list_M = [0,  25, 33, 40, 50, 60, 72, 100]
label_split_M = []

for i in range(len(split_list_M)-1):
    label_split_M.append(str(split_list_M[i]) + "_" + str(split_list_M[i+1]))
```

```python
label_split_M
```

```python
train["Age_cls_M"] = pd.cut(train['Age'], split_list_M, labels=label_split_M)
test["Age_cls_M"] = pd.cut(test['Age'], split_list_M, labels=label_split_M)
```

### 女性用区分

```python
split_list_F = [0, 34, 68, 100]
label_split_F = []

for i in range(len(split_list_F)-1):
    label_split_F.append(str(split_list_F[i]) + "_" + str(split_list_F[i+1]))
```

```python
label_split_F
```

```python
train["Age_cls_F"] = pd.cut(train['Age'], split_list_F, labels=label_split_F)
test["Age_cls_F"] = pd.cut(test['Age'], split_list_F, labels=label_split_F)
```

```python
# 女性データ
for i in train["Age_cls_F"].unique():
    print(i)
    display(train[(train["Gender"]==0)&(train["Age_cls_F"]==i)].corr())
    display(train[(train["Gender"]==0)&(train["Age_cls_F"]==i)].corr()[abs(train[(train["Gender"]==0)&(train["Age_cls_F"]==i)].corr()) > 0.7])
```

```python
# Alb 削除
features_F = ['T_Bil', 'D_Bil', 'ALP', 'ALT_GPT', 'AST_GOT', 'TP', 'AG_ratio']

# 女性用 ALT_GPT AG_ratio 削除
features_F_68_100 = ['T_Bil', 'D_Bil', 'ALP', 'AST_GOT', 'TP', 'Alb']
```

### データの分割

```python
# testデータに復元用のNo.を振る
serial_num = pd.RangeIndex(start=1, stop=test.shape[0]+1, step=1)
test['No'] = serial_num
```

```python
params = {
    'boosting_type': 'gbdt',
    'metric': 'auc',
    'objective': 'binary',
    'verbosity': -1,
    'seed': 42,
    'learning_rate': 0.1,
    'bagging_fraction': 0.75,
    'bagging_freq': 10, 
    'colsample_bytree': 0.75,
}
```

### 年代と性別で区分

```python
test_pred = train.head(0)
test_pred
```

```python
vali_list = {}
```

```python
# 年代別に学習
# 男性のみ
for i in range(1, 2):

    for j in label_split_M:
        train_tmp = train[train["Gender"]==i]
        test_tmp = test[test["Gender"]==i]

        train_tmp = train_tmp[train_tmp["Age_cls_M"]==j]
        test_tmp = test_tmp[test_tmp["Age_cls_M"]==j]
    
        # ホールドアウト法
        X_train, X_test, y_train, y_test = train_test_split(train_tmp[features_M],
                                                            train_tmp["disease"], train_size=0.7, test_size=0.3, random_state=42)

        # lightGBM用のデータセット作成
        train_set = lgb.Dataset(X_train, y_train)
        val_set = lgb.Dataset(X_test, y_test, reference=train_set)
    
        # 学習
        model_lgb = lgb.train(params, train_set, valid_sets=val_set,
                      verbose_eval=None,  # 50イテレーション毎に学習結果出力
                      num_boost_round=100,  # 最大イテレーション回数指定
                      early_stopping_rounds=100
                     )
    
        # 評価
        val_prob = model_lgb.predict(X_test, num_iteration=model_lgb.best_iteration)
        val_pred = np.round(val_prob)
    
        # 誤差
        print(f"性別{i}、{j}歳代")
        print(f"正解率: {accuracy_score(y_test, val_pred)}")
        print(f"適合率: {precision_score(y_test, val_pred)}")
        print(f"再現率: {recall_score(y_test, val_pred)}")
        print(f"f1_score: {f1_score(y_test, val_pred)}\n")
    
        vali_list.update([(str(i)+" "+j, f1_score(y_test, val_pred))])
    
        # 予測データ作成
        y_prob = model_lgb.predict(test_tmp[features_M], num_iteration=model_lgb.best_iteration)
        y_pred = np.round(y_prob)
    
        test_tmp["disease"] = y_pred
        test_pred = pd.concat([test_pred, test_tmp]).sort_values('No')
```

```python
for i in label_split_F:

    train_tmp = train[train["Gender"]==0]
    test_tmp = test[test["Gender"]==0]
    
    train_tmp = train_tmp[train_tmp["Age_cls_F"]==i]
    test_tmp = test_tmp[test_tmp["Age_cls_F"]==i]
    
    print(train_tmp.shape)
    print(test_tmp.shape)
```

```python
# 女性のみ
for i in label_split_F:

    train_tmp = train[train["Gender"]==0]
    test_tmp = test[test["Gender"]==0]
    
    train_tmp = train_tmp[train_tmp["Age_cls_F"]==i]
    test_tmp = test_tmp[test_tmp["Age_cls_F"]==i]
    
    # ホールドアウト法
    if i == "68_100":
        X_train, X_test, y_train, y_test = train_test_split(train_tmp[features_F_68_100],
                                                            train_tmp["disease"], train_size=0.7, test_size=0.3, random_state=42)
    else:
        X_train, X_test, y_train, y_test = train_test_split(train_tmp[features_F],
                                                            train_tmp["disease"], train_size=0.7, test_size=0.3, random_state=42)

    # lightGBM用のデータセット作成
    train_set = lgb.Dataset(X_train, y_train)
    val_set = lgb.Dataset(X_test, y_test, reference=train_set)
    
    # 学習
    model_lgb = lgb.train(params, train_set, valid_sets=val_set,
                    verbose_eval=None,  # 50イテレーション毎に学習結果出力
                    num_boost_round=30,  # 最大イテレーション回数指定
                    early_stopping_rounds=30
                    )
    
    # 評価
    val_prob = model_lgb.predict(X_test, num_iteration=model_lgb.best_iteration)
    val_pred = np.round(val_prob)
    
    # 誤差
    print(f"{i}歳代")
    print(f"正解率: {accuracy_score(y_test, val_pred)}")
    print(f"適合率: {precision_score(y_test, val_pred)}")
    print(f"再現率: {recall_score(y_test, val_pred)}")
    print(f"f1_score: {f1_score(y_test, val_pred)}\n")
    
    vali_list.update([(str(0)+" "+str(i), f1_score(y_test, val_pred))])
    
    # 予測データ作成
    if i == "68_100":
        y_prob = model_lgb.predict(test_tmp[features_F_68_100], num_iteration=model_lgb.best_iteration)
        y_pred = np.round(y_prob)
    else:
        y_prob = model_lgb.predict(test_tmp[features_F], num_iteration=model_lgb.best_iteration)
        y_pred = np.round(y_prob)

    test_tmp["disease"] = y_pred
    test_pred2 = pd.concat([test_pred2, test_tmp]).sort_values('No')
```

### ホールドアウト法

```python
X_train, X_test, y_train, y_test = train_test_split(train[features], 
                                                    train["disease"], test_size=0.3, random_state=42)
```

### 学習

#### lightGBM

```python
train_set = lgb.Dataset(X_train, y_train)
val_set = lgb.Dataset(X_test, y_test, reference=train_set)
```

### 1

```python
params42 = {
    'boosting_type': 'gbdt',
    'metric': 'auc',
    'objective': 'binary',
    'verbosity': -1,
    'seed': 42,
    'learning_rate': 0.1,
    'bagging_fraction': 0.75,
    'bagging_freq': 10, 
    'colsample_bytree': 0.75,
}
```

```python
model_lgb42 = lgb.train(params, train_set, valid_sets=val_set,
                  verbose_eval=50,  # 50イテレーション毎に学習結果出力
                  num_boost_round=1000,  # 最大イテレーション回数指定
                  early_stopping_rounds=100
                 )
```

```python
val_prob42 = model_lgb42.predict(X_test, num_iteration=model_lgb42.best_iteration)
val_pred42 = np.round(val_prob42)
```

```python
# 誤差
print(f"正解率: {accuracy_score(y_test, val_pred42)}")
print(f"適合率: {precision_score(y_test, val_pred42)}")
print(f"再現率: {recall_score(y_test, val_pred42)}")
print(f"f1_score: {f1_score(y_test, val_pred42)}")
```

```python
y_prob42 = model_lgb42.predict(test[features], num_iteration=model_lgb42.best_iteration)
y_pred42 = np.round(y_prob42)
```
